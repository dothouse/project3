```python
# .zip 파일 리스트
datadir = '/content/drive/MyDrive/project3/data/traindata/raw_data/'
zip_files = [f for f in os.listdir(datadir) if f.endswith('.zip') and f != 'kfood.zip']

# 압축 해제 함수
def unzip_file(zip_file_path, extract_to):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# 모든 .zip 파일 압축 해제 및 복사
for zip_file in tqdm(zip_files):
    print(zip_file)
    zip_file_path = os.path.join(datadir, zip_file)
    extract_to = '/content/drive/MyDrive/project3/data/traindata/food_image'

    # 압축 해제
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)

    unzip_file(zip_file_path, extract_to)

    # 압축 해제된 파일 - 코랩상에 임시 형태로 존재
    # 이를 Google Drive로 복사
    for item in os.listdir(extract_to):
        start_folder = os.path.join(extract_to, item)
        dest_folder = os.path.join(datadir, item)
        if os.path.isdir(start_folder):
            shutil.copytree(start_folder, dest_folder, dirs_exist_ok=True)
        else:
            shutil.copy2(start_folder, dest_folder)

    # 압축 해제된 임시 파일들 삭제
    shutil.rmtree(extract_to)

print("압축 해제 및 파일 복사 완료!")
```

```python
# 데이터 경로 지정

datadir = '/content/drive/MyDrive/project3/data/traindata/food_image/'

food_middle_list = sorted(os.listdir(datadir))
len(food_middle_list)
```

```python
#  데이터 확인

food_list = [] # 음식이름
food_path_list = [] # 음식이름 경로

for food_middle in food_middle_list:
    middle_path = os.path.join(datadir,food_middle)
    food_name_list = os.listdir(middle_path)

    for food in food_name_list:
        path = os.path.join(middle_path,food)
        food_path_list.append(path)

        name = path.split('/')[-1]
        food_list.append(name)

print(len(food_list))
print(food_list)
print(food_path_list)
```
