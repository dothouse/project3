```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_excel('/content/drive/MyDrive/project3/data/practice_text/recipe_song/만개의 레시피1.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/project3/data/practice_text/recipe_song/만개의 레시피2.xlsx')
metadata = pd.read_excel('/content/drive/MyDrive/project3/data/practice_text/recipe_song/만개의 레시피_데이터명세서.xlsx')
```

```python
print(df1.shape)
print(df2.shape)
print(metadata.shape)
```

```python
metadata.head()
```

```python
recipe_dict = dict(zip(metadata['nameEn'], metadata['nameKo']))
recipe_dict
```

```python
df1.rename(columns=recipe_dict, inplace=True)
df2.rename(columns=recipe_dict, inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2], axis=0)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['temp'] = df['요리재료내용'].apply(lambda x: x.replace('[재료]',"").split('|'))
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/project3/data/practice_text/recipe_song/recipe.csv', index=False)
```
