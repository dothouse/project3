```python
import pandas as pd
import numpy as np

ingr=pd.read_excel("./요리명+요리재료내용.xlsx")
ingr.head()
```

```python
import re

# 정규 표현식을 사용하여 '[양념재료]', '[양념]', '[양념장]' 이후의 모든 문자열을 삭제하여 새로운 컬럼에 할당
ingr['필수재료'] = ingr['요리재료내용'].apply(lambda x: re.sub(r'\[(양념재료|양념|양념장)\].*', '', x))

# 새로운 데이터프레임 확인
print(ingr[['요리재료내용', '필수재료']])
```

```python
ingr
```

```python
# '필수재료' 컬럼에서 '|'를 ','로 변경
ingr['필수재료'] = ingr['필수재료'].str.replace('|', ',')
ingr.head()
```

```python
# '필수재료' 컬럼 내 숫자부터 쉼표(,) 전까지의 문자열 제거
ingr['필수재료'] = ingr['필수재료'].str.replace(r'\d+[^,]*', '', regex=True)
```

```python
# '필수재료' 컬럼에서 '또는'을 'or'로 변경
ingr['필수재료'] = ingr['필수재료'].str.replace('또는', 'or')
ingr.head()
```

```python
# '필수재료' 컬럼에서 '/'를 'or'로 변경
ingr['필수재료'] = ingr['필수재료'].str.replace('/', 'or')
ingr.head()
```

```python
# '필수재료' 컬럼 내 '?' 문자열 삭제
ingr['필수재료'] = ingr['필수재료'].str.replace('?', '', regex=False)
```

```python
# '필수재료' 컬럼 내 '[]' 안의 값을 모두 삭제
ingr['필수재료'] = ingr['필수재료'].str.replace(r'\[[^\]]*\]', '', regex=True)
ingr.head()
```

```python
# '필수재료' 컬럼 내 '()' 안의 값을 모두 삭제
ingr['필수재료'] = ingr['필수재료'].str.replace(r'\([^)]*\)', '', regex=True)
ingr.head()
```

```python
# '필수재료' 컬럼 내 '양념장' 제거
ingr['필수재료'] = ingr['필수재료'].str.replace('양념장', '', regex=False)
ingr.head()
```

```python
# '대파나 쪽파'를 '대파'로 변경
ingr['필수재료'] = ingr['필수재료'].str.replace('대파나 쪽파', '대파', regex=False)
```

```python
# '필수재료' 컬럼 내 'or'로 구분된 문자열에서 첫 번째 값만 남기기
ingr['필수재료'] = ingr['필수재료'].str.split('or').str[0]
ingr.head()
```

```python
# '필수재료' 컬럼 내 쉼표(,)를 제외한 모든 특수 문자를 공백으로 대체
ingr['필수재료'] = ingr['필수재료'].str.replace(r'[^\w\s,]', ' ', regex=True)

# 쉼표(,)를 제외한 모든 공백을 제거하여 쉼표(,)만 남기기
ingr['필수재료'] = ingr['필수재료'].str.replace(r'\s+', '', regex=True)
ingr.head()
```

```python
# '젖'을 '젓'으로 변경
ingr['필수재료'] = ingr['필수재료'].str.replace('젖', '젓', regex=False)
```

```python
# '필수재료' 컬럼 내 주어진 문구들을 모두 빈 문자열로 대체하여 제거
to_remove = ['입맛에 맞게', '적당량', '조금', '원하는만큼', '천일염 하는양에따라', '넉넉히', '솔솔', '약간', '큰거', '큰것', '큰걸로', '작은것', '작은거', '작은크기',
             '작은 크기', '취향에 따라', '작은술', '작은 사이즈', '크게', '적당히', '톡톡', '톡톡톡물', '소복히', '엄지손가락 반만한크기', '취향껏', '팍팍', '마지막간',
             '마무리용', '마무리양념', '원하는양', '원하는만큼', '원하는 만큼', '소량', '가득', '두르기']
for phrase in to_remove:
    ingr['필수재료'] = ingr['필수재료'].str.replace(phrase, '', regex=False)

ingr    
```

```python
ingr.to_csv("./필수재료 생성.csv",index=False)
```

```python
# '요리재료내용' 컬럼 내 ',,'를 ','로 수정
ingr['요리재료내용'] = ingr['요리재료내용'].str.replace(r',+', ',', regex=True)

# '요리재료내용' 컬럼 내 'or' 뒤의 내용을 삭제
ingr['요리재료내용'] = ingr['요리재료내용'].apply(lambda x: x.split('or')[0])

# '요리재료내용' 컬럼 내 쉼표(,)를 ', '로 수정
ingr['요리재료내용'] = ingr['요리재료내용'].str.replace(',', ', ', regex=False)
```

```python
from konlpy.tag import Okt

# Okt 객체 생성
okt = Okt()

# 특정 컬럼을 형태소와 품사를 함께 추출하여 새로운 컬럼에 추가
ingr['요리재료내용_Okt'] = ingr['요리재료내용'].apply(lambda x: okt.pos(x))
ingr.head()
```

```python
# 명사(N으로 시작하는 품사)만 추출하는 함수 정의
def extract_nouns(pos_list):
    nouns = [word for word, pos in pos_list if pos.startswith('N')]
    return nouns

# '요리재료내용_Okt' 컬럼을 이용하여 명사만 추출하여 새로운 컬럼에 추가
ingr['Nouns_Okt'] = ingr['요리재료내용_Okt'].apply(extract_nouns)
ingr.head(10)
```

```python
from konlpy.tag import Komoran

# Komoran 객체 생성
komoran=Komoran()

# 특정 컬럼을 형태소와 품사를 함께 추출하여 새로운 컬럼에 추가
ingr['요리재료내용_Komoran'] = ingr['요리재료내용'].apply(lambda x: komoran.pos(x))
ingr.head()
```

```python
# 명사(N으로 시작하는 품사)만 추출하는 함수 정의
def extract_nouns(pos_list):
    nouns = [word for word, pos in pos_list if pos.startswith('N')]
    return nouns

# '요리재료내용_Komoran' 컬럼을 이용하여 명사만 추출하여 새로운 컬럼에 추가
ingr['Nouns_Komoran'] = ingr['요리재료내용_Komoran'].apply(extract_nouns)
ingr.head(10)
```
