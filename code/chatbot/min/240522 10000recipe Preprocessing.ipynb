```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./recipe_data.csv")
df
```

```python
# 컬럼명 변경
df.rename(columns={'step':'조리순서'}, inplace=True)
df.head()
```

```python
df['조리순서'] = df['조리순서'].str.replace("[", "")
df
```

```python
df['조리순서'] = df['조리순서'].str.replace("]", "")
df
```

```python
df['조리순서'] = df['조리순서'].str.replace("'", "")
df
```

```python
#  '조리순서' 컬럼의 각 값에서 '\n' 제거
df['조리순서'] = df['조리순서'].str.replace(r'\n', '')
df
```

```python
# '조리순서' 컬럼의 값에서 ':'을 '.'으로 변경하고 숫자 앞에 '\n' 추가
def replace_step_with_numbers(text):
    step_number = 1
    new_text = ""
    for word in text.split():
        if word.startswith("step:"):
            if step_number != 1:
                new_text += "\n"
            new_text += str(step_number) + "."
            step_number += 1
            new_text += word[5:] + " "
        else:
            new_text += word + " "
    return new_text.strip()

df['조리순서'] = df['조리순서'].apply(replace_step_with_numbers)
df
```

```python
# '[조리순서]' 다음에 줄바꿈 추가
df['조리순서'] = df['조리순서'].apply(lambda x: '[조리순서]' + '\n' + x)
df
```

```python
df['조리순서'] = df['조리순서'].str.replace(', \n', '\n')
df.head()
```

```python
df.to_csv("./10000recipe Preprocessing.csv", index=False)
```
