```python
ingr=pd.read_excel("./요리명+요리재료내용.xlsx")
ingr.head()
```

```python
# 정규식 패턴 설정
pattern_ingredient = r'\[재료\]([^[]*?)(?=\[|$)'
pattern_required_ingredient = r'\[필수재료\]([^[]*?)(?=\[|$)'
```

```python
def extract_ingredients(ingr):
    if not isinstance(ingr, str):
        return ''  # 입력이 문자열이 아닌 경우 빈 문자열 반환
    
    match_required_ingredient = re.search(pattern_required_ingredient, ingr)
    if match_required_ingredient:
        extracted_required_ingredients = match_required_ingredient.group(1).strip()
        return extracted_required_ingredients
    else:
        match_ingredient = re.search(pattern_ingredient, ingr)
        if match_ingredient:
            extracted_ingredients = match_ingredient.group(1).strip()
            return extracted_ingredients
        else:
            return ''



```

```python
ingr['재료'] = ingr['요리재료내용'].apply(extract_ingredients)
```

```python
ingr
```

```python
# '재료' 컬럼에서 '|'를 ','로 변경
ingr['재료'] = ingr['재료'].str.replace('|', ',')
ingr.head()
```

```python
import re

# '재료' 컬럼 내 숫자부터 쉼표(,)까지의 값들을 제거
ingr['재료'] = ingr['재료'].str.replace(r'\d[^,]*,', ',', regex=True)
ingr.head()
```

```python
# '재료' 컬럼 내 숫자와 숫자 뒤의 문자열을 제거
ingr['재료'] = ingr['재료'].str.replace(r'\d.*$', '', regex=True)
ingr.head()
```

```python
# '요리재료내용' 컬럼에서 '또는'을 'or'로 변경
ingr['재료'] = ingr['재료'].str.replace('또는', 'or')
ingr.head()
```

```python
# '재료' 컬럼에서 '/'를 'or'로 변경
ingr['재료'] = ingr['재료'].str.replace('/', 'or')
ingr.head()
```

```python
# '재료' 컬럼 내 '?' 문자열 삭제
ingr['재료'] = ingr['재료'].str.replace('?', '', regex=False)
```

```python
# '요리재료내용' 컬럼 내 ' ,'를 찾아서 공백을 없애고 쉼표(,)만 남기기
ingr['재료'] = ingr['재료'].str.replace(r'\s*,\s*', ', ', regex=True)
ingr.head()
```

```python
# '재료' 컬럼 내 '()' 안의 값을 모두 삭제
ingr['재료'] = ingr['재료'].str.replace(r'\([^)]*\)', '', regex=True)
ingr.head()
```

```python
# '재료' 컬럼 내 쉼표(,)를 제외한 모든 특수 문자를 공백으로 대체
ingr['재료'] = ingr['재료'].str.replace(r'[^\w\s,]', ' ', regex=True)

# 쉼표(,)를 제외한 모든 공백을 제거하여 쉼표(,)만 남기기
ingr['재료'] = ingr['재료'].str.replace(r'\s+', '', regex=True)
ingr.head()
```

```python
ingr['재료'] = ingr['재료'].str.replace(r'\s*or\s*.*$', '', regex=True)
```
