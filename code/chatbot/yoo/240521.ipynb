```python
import pandas as pd
```

```python
reci = pd.read_csv('./만개의레시피수정.csv')
reci.head()
```

```python
# rule의 데이터를 split하여 list 형태로 변환 후, index값과 함께 dictionary 형태로 저장

chat_dic={} # 빈 딕셔너리 초기화
row=0 # 'row'라는 변수를 0으로 초기화

for rule in reci['CKG_NM']:
    chat_dic[row]=rule.split('|') # 각 rule 항목을 '|' 문자를 기준으로 분할하여 리스트로 변환
    row += 1 # row 값을 1씩 증가시킴

print(chat_dic)
```

```python
def chat(request): # 'request'라는 하나의 매개변수를 받음
    for k, v in chat_dic.items():
        chat_flag=False # 'chat_flag' 변수를 'False'로 초기화
        # 내부 반복문 및 조건문
        for word in v:
            if word in request:
                chat_flag=True
            else:
                chat_flag=False
                break
            # 응답 반환
            if chat_flag:
                return reci['CKG_NM'][k] + '\n' + reci['CKG_MTRL_CN'][k] + '\n' + reci['Column4'][k]
            
    return '음식명으로 검색해주세요.. 데헷'

# 'chat_flag'가 False이면 '무슨 말인지 잘 모르겠구나..!'라는 문자열을 반환   
```

```python
while True:
    req=input("채팅을 입력하세요!")
    if req == '끝':
        break
    else:
        print("쿠킹파트너 : ",chat(req))

# 만들어진 챗봇 함수를 호출하여 rule이 모두 포함되어 있으면 response를 반환, 프로그램을 종료하고 싶으면 'exit' 입력
```
