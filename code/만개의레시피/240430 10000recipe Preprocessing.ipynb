```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./recipe.csv", encoding='utf-8')
df
```

```python
import re

string = df['재료']
new_string = string.apply(lambda x: re.sub(r'[재료]', '', x))
print(new_string)
```

```python
# '[재료]'를 빈 문자열로 대체
df['재료'] = df['재료'].apply(lambda x: x.replace('[재료]', ''))
df
```

```python
# '\n구매'를 빈 문자열로 대체
df['재료'] = df['재료'].apply(lambda x: x.replace('\n구매', ''))

df
```

```python
df['재료'] = df['재료'].apply(lambda x: x.lstrip('\n'))
df
```

```python
df['재료'] = df['재료'].str.replace('\n', ' ')
df
```

```python
df['조리방법'] = df['조리방법'].str.replace("[", "")
df
```

```python
df['조리방법'] = df['조리방법'].str.replace("'", "")
df
```
