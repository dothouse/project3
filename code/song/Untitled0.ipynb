```python
# 파일 경로
file_path = '/content/drive/MyDrive/project3/data/traindata/kfood/국/계란국/crop_area.properties'

# 딕셔너리 초기화
crop_areas = {}

# 파일 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # 각 줄에서 키-값 쌍 분리
        key, value = line.strip().split('=')
        # 값을 정수 리스트로 변환
        coordinates = list(map(int, value.split(',')))
        # 딕셔너리에 저장
        crop_areas[key] = coordinates

# 결과 출력
for key, value in crop_areas.items():
    print(f"{key}: {value}")
```

```python
import cv2
import os

# 파일 경로
properties_file_path = '/content/drive/MyDrive/project3/data/traindata/kfood/나물/미역줄기볶음/crop_area.properties'
images_folder_path = '/content/drive/MyDrive/project3/data/traindata/kfood/나물/미역줄기볶음/'
output_folder_path = '/content/drive/MyDrive/project3/data/traindata/kfood_box/나물/미역줄기볶음/boxed/'


# 딕셔너리 초기화
crop_areas = {}

# 폴더가 존재하지 않으면 생성
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# 파일 읽기
with open(properties_file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # 각 줄에서 키-값 쌍 분리
        key, value = line.strip().split('=')

        # 값을 정수 리스트로 변환
        coordinates = list(map(int, value.split(',')))
        crop_areas[key] = coordinates

        # 이미지 파일 경로
        image_path = os.path.join(images_folder_path, f"{key}.jpg")
        if not os.path.exists(image_path):
            image_path = os.path.join(images_folder_path, f"{key}.png")

        # 이미지 읽기
        if os.path.exists(image_path):
            image = cv2.imread(image_path)
            if image is not None:
                # 박스 그리기 (x, y, w, h)
                x, y, w, h = coordinates
                boxed_image = cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

                # 박스를 그린 이미지 저장
                output_image_path = os.path.join(output_folder_path, f"boxed_{key}.jpg")
                cv2.imwrite(output_image_path, boxed_image)
                print(f"Boxed image saved: {output_image_path}")
            else:
                print(f"Failed to read image: {image_path}")
        else:
            print(f"Image file not found: {image_path}")
```
