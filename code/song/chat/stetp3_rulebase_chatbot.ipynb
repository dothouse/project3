```python
import pandas as pd
import numpy as np

import os
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/drive/MyDrive/project3/data/image_chatbot/recipe_final.csv')
df.head()
```

```python
df_chat = df.copy()

food_name_list = df_chat['요리명'].unique()


while True:
    request = input('알고 싶은 요리명을 알려주세요. 종료하려면 끝을 입력하세요')

    temp = df_chat[df_chat['요리명'] == request].head(1)
    if request in food_name_list:
        print(request, " 조리법은?")
        print(temp['step'].values[0])
        step2 = input('재료를 알려드릴까요? (네 / 아니요')
        if step2 == '네':
            print(temp['재료'].values[0])
        else :
            continue

    elif request ==  '끝':
        break
    else :
        print('요리명을 정확하게 다시 입력하세요')
```
