```python
import pandas as pd
import numpy as np

import re

import time
#time.sleep을 위한 난수생성
import random


import os
import warnings
warnings.filterwarnings("ignore")
```

# Crawling data


```python
datadir = '/content/drive/MyDrive/project3/data/image_chatbot/steps/'

file_list = os.listdir(datadir)
file_list = sorted(file_list)
selected_csv =[csv for csv in file_list if csv.split('.')[-1] == 'csv']
selected_csv
```

```python
df_step = pd.DataFrame(columns = ['RCP_SNO', 'step'])

for file_name in selected_csv:
    folder_path = datadir
    path = folder_path + file_name

    temp = pd.read_csv(path)

    df_step = pd.concat([df_step, temp], axis = 0)
```

```python
df_step.head()
```

# Download data + Crwaling data


```python
df = pd.read_excel("/content/drive/MyDrive/project3/data/image_chatbot/만개의 레시피1.xlsx")
df.head()
```

```python
df_final = df_step.merge(df,
                         how = 'left',
                         on = 'RCP_SNO')

df_final.head()
```

```python
metadata = pd.read_excel('/content/drive/MyDrive/project3/data/image_chatbot/만개의 레시피_데이터명세서.xlsx')
metadata.head()
```

```python
recipe_dict = dict(zip(metadata['nameEn'], metadata['nameKo']))
recipe_dict
```

```python
df_final = df_step.merge(df,
                         how = 'left',
                         on = 'RCP_SNO')


df_final.rename(columns=recipe_dict, inplace=True)
df_final.dropna(inplace=True)

df_final.head()
```

```python
df_final['재료'] = df_final['요리재료내용'].apply(lambda x: re.sub(r'\[[가-힇]+\]', '|', x))
df_final['재료'] = df_final['재료'].apply(lambda x : x.replace(' ', '').replace('|', ' '))
df_final.head()
```
